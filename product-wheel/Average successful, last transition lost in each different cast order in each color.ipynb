```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('../data/OrderColorTransitionsReport 1_1_17 - 4_22_19.xlsx',
              header=205, usecols=[0,1,4,5,7,8,9,11,12])
```

```python
data.loc[:, 'Cast Start'] = pd.to_datetime(data.loc[:, 'Cast Start'])
data.loc[:, 'Cast End Time'] = pd.to_datetime(data.loc[:, 'Cast End Time'])
data['Duration'] = data['Cast End Time'] - data['Cast Start']
data
```

```python
def Select(color1, color2):
    color = data['Color'] == color1
    nextcolorcode = data ['Next Color Code'] == color2
    return data[color & nextcolorcode]
```

```python
Select('1M', 'CW')
```

```python
def SelectCalc(color1, color2):
    color = data['Color'] == color1
    nextcolorcode = data ['Next Color Code'] == color2
    Selecteddf = data[color & nextcolorcode]
    n = 0
    b = 0
    for i in range(len(Selecteddf)-1):
        if Selecteddf.iloc[i,0] == Selecteddf.iloc[i+1,0]:
            n = n + 1
        else: 
            b = b + 1
    s = s + int(Selecteddf.iloc[n,8])
    x = x + int(Selecteddf.iloc[b+n,8])
    Average = (s+x)/2
    print ('First Batch', n+1)
    print ('Second Batch #', b)
    print ('last row batch 1 yo', s)
    print ('last row batch 2 yo', x)
    print ('Average', Average)
```

```python
SelectCalc('1M', 'CW')
```
